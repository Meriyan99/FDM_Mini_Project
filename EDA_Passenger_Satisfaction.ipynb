# Airline Passenger Satisfaction Dataset

## Dataset Information
1. Gender: Gender of the passengers (Female, Male)
2. Customer Type: The customer type (Loyal customer, disloyal customer)
3. Age: The actual age of the passengers
4. Type of Travel: Purpose of the flight of the passengers (Personal Travel, Business Travel)
5. Class: Travel class in the plane of the passengers (Business, Eco, Eco Plus)
6. Flight distance: The flight distance of this journey
7. Inflight wifi service: Satisfaction level of the inflight wifi service (0:Not Applicable;1-5)
8. Departure/Arrival time convenient: Satisfaction level of Departure/Arrival time convenient
9. Ease of Online booking: Satisfaction level of online booking
10. Gate location: Satisfaction level of Gate location
11. Food and drink: Satisfaction level of Food and drink
12. Online boarding: Satisfaction level of online boarding
13. Seat comfort: Satisfaction level of Seat comfort
14. Inflight entertainment: Satisfaction level of inflight entertainment
15. On-board service: Satisfaction level of On-board service
16. Leg room service: Satisfaction level of Leg room service
17. Baggage handling: Satisfaction level of baggage handling
18. Check-in service: Satisfaction level of Check-in service
19. Inflight service: Satisfaction level of inflight service
20. Cleanliness: Satisfaction level of Cleanliness
21. Departure Delay in Minutes: Minutes delayed when departure
22. Arrival Delay in Minutes: Minutes delayed when Arrival

## Data Visualization

```python
# this code segment will seperate categorical columns from the entire dataset and describe each column by mentioning 
# number of unique values of each column, top unique value for each column and their frequencies

categorical_passenger_data = [0,1,3,4] + list(range(6,20))
passenger_data.iloc[:,categorical_passenger_data] = passenger_data.iloc[:,categorical_passenger_data].astype('category')
passenger_data.describe(include='category')
```

```python
#plot the output column
plt.pie(passenger_data.satisfaction.value_counts(),colors = sns.color_palette("mako"),labels=['neutral or dissatisfied','satisfied'],autopct = '%1.1f%%')
```

#### Pie chart shows for both categories in the output column the selection is more or less balanced 

### Let's plot each categorical column to see the distribution of each unique values present in categorical columns

```python
category = passenger_data.iloc[:,categorical_passenger_data]
fig, axes = plt.subplots(6, 3, figsize = (15, 15))
for i, col in enumerate(category):
    column_values = passenger_data[col].value_counts()
    labels = column_values.index
    sizes = column_values.values
    axes[i//3, i%3].pie(sizes, labels = labels, colors = sns.color_palette("mako"), autopct = '%1.0f%%', startangle = 90)
    axes[i//3, i%3].axis('equal')
    axes[i//3, i%3].set_title(col)
plt.show()
```

### Conclusions
1. Male and Female percentages are almost same
2. Majority of the customer type is loyal
3. Majority of the passengers are tend to business travel rather personal travel
4. Nearly 50% of the passengers are from business class
5. Satisfaction level of baggage handling,inflight services,leg room service,seat comfort,on-board service,inflight entertainment are rated by majority as 4-5 

#### There is a high correlation between Departure Delay in Minutes and Arrival Delay in Minutes columns

```python
plt.scatter(passenger_data['Arrival Delay in Minutes'],passenger_data['Departure Delay in Minutes'],alpha=0.5)
```

#### Conclusion: Dependance between Departure Delay in Minutes and Arrival Delay in Minutes are linear. This indicates that if flight delayed by a certain amount of time at the depature, there is a high chance of delay the same flight by the same amount of time at the landing as well
